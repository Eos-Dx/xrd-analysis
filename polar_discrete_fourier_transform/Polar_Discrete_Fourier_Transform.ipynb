```python
"""
Import necessary libraries
"""
import os
import glob

import numpy as np
from scipy.special import jn_zeros
from scipy.io import savemat
import matplotlib.pyplot as plt

from skimage.transform import warp_polar

from preprocessing.preprocess import PreprocessData
from models.utils import gen_zeromatrix
```

```python
"""
Generate or load zeromatrix file
"""
nthorder = 10
kzeros = 10
# zeromatrix = gen_zeromatrix((nthorder,kzeros),save_mat=True,save_numpy=True,outdir="")
zeromatrix = np.load(os.path.join("","zeromatrix.npy"))
```

```python
"""
Load a normal and a cancer file
normal: A00061
cancer: A00060
"""
datadir = "discrete_polar_fourier_transform/data/"

normal_data = np.loadtxt(os.path.join(datadir,"preprocessed_A00030.txt"))
cancer_data = np.loadtxt(os.path.join(datadir,"preprocessed_A00053.txt"))

plt.imshow(normal_data)
plt.show()
plt.imshow(cancer_data)
plt.show()
```

```python
"""
Warp polar
Transform cartesian to polar coordinates
"""
radius=85

# Warp polar
normal_polar = warp_polar(normal_data,radius=radius, preserve_range=True).round()
cancer_polar = warp_polar(cancer_data,radius=radius, preserve_range=True).round()

print(normal_polar.shape)
print(cancer_polar.shape)

print(np.max(normal_polar))
print(np.max(cancer_polar))

fig = plt.figure(figsize=(8,16))
fig.set_facecolor("white")
ax1 = fig.add_subplot(121)

ax1.imshow(normal_polar, vmax=None)
ax1.set_aspect(0.5)
plt.title("Normal: " + "preprocessed_A00030")

ax2 = fig.add_subplot(122)
ax2.imshow(cancer_polar, vmax=None)
ax2.set_aspect(0.5)
plt.title("Cancer: " + "preprocessed_A00053")

plt.show()
```
